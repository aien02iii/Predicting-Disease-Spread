## 4.predict

```python
regr = linear_model.LinearRegression()

new_data = data4[[
                  'weekofyear',
#                   'season',
#                   'year',
                  'month',
#                   'two_mon',
#                   'city_num',
#                   'TatalRain_80',
                  'Temp_stitu_300',
                  'ndvi_ne2',
                  'ndvi_nw2', 
                  'ndvi_se2',
                  'ndvi_sw2',   
                  'precipitation_amt_mm2',
                  'reanalysis_air_temp_k2',
                  'reanalysis_avg_temp_k2',
    
                  'reanalysis_dew_point_temp_k2',
                  'reanalysis_max_air_temp_k2',    
                  'reanalysis_min_air_temp_k2',
                  'reanalysis_precip_amt_kg_per_m2_2',
                  'reanalysis_relative_humidity_percent2',
    
                  'reanalysis_sat_precip_amt_mm2',
#                   'reanalysis_specific_humidity_g_per_kg2',
                  'reanalysis_tdtr_k2',    
                  'station_avg_temp_c2',
#                   'station_diur_temp_rng_c2',
                  'station_max_temp_c2',
#                   'station_min_temp_c2', 
                  'station_precip_mm2'
                ]]
new_data
```

```python
X = new_data.values
y = data4.total_cases.values
```

### 特徵標準化(StandardScaler)、主成分分析(PCA)

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y ,test_size=0.3, random_state=123)
```

```python
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)

pca = PCA(n_components=17)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)

# regr.fit(X_train_std, y_train)
# y_pred = regr.predict(X_test_std)
# print(y_pred)
# print("=="*50)
# print("Score:"+str(regr.score(X_test_std, y_test)))
# print("MAE:"+str(mean_absolute_error(y_test, y_pred)))
# print("MSE:",metrics.mean_squared_error(y_test, y_pred))
# print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))


regr.fit(X_train_pca, y_train)
y_pred = regr.predict(X_test_pca)
print(y_pred)
print("=="*50)
print("Score:"+str(regr.score(X_test_pca, y_test)))
print("MAE:"+str(mean_absolute_error(y_test, y_pred)))
print("MSE:",metrics.mean_squared_error(y_test, y_pred))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
