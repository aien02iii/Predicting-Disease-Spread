```python
#實務應用1:groupby + apply

series3 = pd.Series([60, 70, 80, 90, 120, 130])
def myfunc2(a):
    if a >= 100:
        return 'Vip'
    elif a >= 80 and a < 100:
        return 'Ip'
    else:
        return 'P'

s3 = series3.apply(myfunc2)
g3 = series3.groupby(s3)
print(g3.count())
print(g3.mean())

#實務應用2:groupby + apply
#strp1:建立series
series3 = pd.Series([60, 71, 80, 91, 120, 131])
#step2:使用apply設定分群條件及群組名稱
s4 = series3.apply(lambda x : '偶數' if x % 2 != 0 else '奇數')
#step3:分群
g4 = series3.groupby(s4)
#step4:分群後的計算
print(g4.count())
print(g4.mean())
print(g4.value_counts())

#唯一、單一值:unique()、計數:value_counts()
series1 = pd.Series([100, 90, 90, 80, 80, 80, 70, 70])
print(series1.unique())
print(series1.count())
print(series1.value_counts())
print(series1.groupby(series1).count())
```

```python
import numpy as np
import pandas as pd


from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.cross_validation import train_test_split
from sklearn.externals import joblib

data = pd.read_csv("Datasets/titanic.csv")
#print(data)

median_age = data['Age'].median()
data['Age'].fillna(median_age, inplace = True)

data_inputs = data[["Pclass", "Age", "Sex"]]
#print(data_inputs)

expected_output = data[["Survived"]]
#print(expected_output)

data_inputs["Pclass"].replace("3rd", 3, inplace = True)
data_inputs["Pclass"].replace("2nd", 2, inplace = True)
data_inputs["Pclass"].replace("1st", 1, inplace = True)
#print(data_inputs)

data_inputs["Sex"] = np.where(data_inputs["Sex"] == "female", 0, 1)
#print(data_inputs)

inputs_train, inputs_test, expected_output_train, expected_output_test = train_test_split (data_inputs, expected_output, test_size = 0.2, random_state = 123)
 
    
#--------------------------1.(建模)執行機器學習-----------------------------
rf = RandomForestClassifier (n_estimators=200, max_depth=20, max_leaf_nodes=5000)
rf.fit(inputs_train, expected_output_train)

#------------------------2-1.(建模)驗證學習結果(test data)---------------------
accuracy = rf.score(inputs_test, expected_output_test)
print("Accuracy = {}%".format(accuracy * 100))

#------------------------2-2.(建模)交叉驗證(不分train、test)---------------------
scores = cross_val_score(rf, data_inputs, expected_output, cv=50)
scores.sort()
accuracy = scores.mean()

print(scores)
print(accuracy)



#------------------------3.(實戰)使用陌生資料驗證(沒有label)---------------------
## test with unseen data
test = pd.read_csv("Datasets/titanic_test.csv")

#median_age = test['Age'].median()
test['Age'].fillna(median_age, inplace = True)

#test["Pclass"].replace("3rd", 3, inplace = True)
test["Pclass"].replace("2nd", 2, inplace = True)
test["Pclass"].replace("1st", 1, inplace = True)
##print(data_inputs)

test["Sex"] = np.where(test["Sex"] == "female", 0, 1)
##print(data_inputs)

test_predictions = rf.predict(test[["Pclass", "Age", "Sex"]])
print(test_predictions.sum())

joblib.dump(rf, "titanic_randomforest_model", compress=9)


****************************************判斷是否OVERFITTING 再回STEP1、STEP2*******************************************
```
